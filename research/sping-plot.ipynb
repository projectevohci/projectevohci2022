# Introduction
The following will results of tests to find the right spring for the kicker  
  
The equation used will be $F=kx$ where $F$ is the force exerted by the spring, $k$ is the spring constant, $x$ is the change in the length of the spring.  
  
The spring used will be an Extension Spring

# Pre-requisties

![alt text](spring-test.jpeg)



```python
import matplotlib.pyplot as plt
import numpy as np

def plotGraph(x,y):
    slope, intercept = np.polyfit(x, y, 1)
    plt.figure()
    plt.scatter(x, y) 
    plt.xlabel("Force / N")
    plt.ylabel("Distance of Extension (m)")
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color = 'k')
    plt.show()

    slope = round(slope, 3)
    slope = str(slope)
    print("Gradient of graph: " + slope)
```

Spring Specifications:  
Wire diameter: 0.4mm  
Diameter of Spring: 4mm  
Length of Spring: 22mm

```python
x = [1, 2, 3, 4, 5, 6]
y = [0.3, 1.2, 2.3, 3.3, 5.8, 6.8]
y = [x/100 for x in y]
x = np.array(x)
y = np.array(y)

plotGraph(x,y)
```
